```python
#@title Get problem data A,b,c
b = jnp.array( [nn[1] for nn in nutrients] )
c = jnp.array( [dd[2] for dd in data] )

A = np.zeros( ( len( nutrients ), len(data) ) )

for j,dd in enumerate( data ):
  A[:,j] = dd[3:]
A = jnp.array( A )
```

```python
#@title Q1) Interior point method - define functions

def barrier_objective( x, mu ):
  return c @ x - mu * jnp.sum( jnp.log(x) ) - mu * jnp.sum( jnp.log(A @ x - b) )

grad_fn = jax.grad(barrier_objective)
hess_fn = jax.hessian(barrier_objective)

def barrier_newton( A, b, c, x0, mu_init=1.0, sigma=0.1, tol=1e-6, max_iter=50 ):
  x = x0
  mu = mu_init
  all_histories = []

  while mu > 1e-8:
    history = []
    for _ in range(max_iter):
      grad = grad_fn(x, mu)
      Hess = hess_fn(x, mu)
      dx = -jnp.linalg.solve(Hess, grad)

      # Backtracking line search for feasibility
      alpha = 1.0
      while jnp.any(x + alpha*dx <= 0) or jnp.any(A @ (x + alpha*dx) - b <= 0):
        alpha *= 0.5
      x = x + alpha * dx
      history.append(float(jnp.linalg.norm(grad)))

      if jnp.linalg.norm(grad) < tol:
        break
      all_histories.append(history)
      mu *= sigma
  return x, all_histories
```

```python
# @title Q1) Solve problem

x0 = jnp.ones( len(data) ) # initial feasible point
x_opt, all_histories = barrier_newton( A, b, c, x0 )

# @title Q1) Plot solution

print("Optimal cost:", c @ x_opt)

# Plot convergence
plt.figure(figsize=(8,5))
for i, hist in enumerate(all_histories):
  plt.semilogy(hist, label=f'Barrier μ-subproblem {i+1}') # semilogy for better visualization
plt.xlabel('Newton iteration')
plt.ylabel('Residual norm ||grad φ(x)||')
plt.title('Convergence of Newton Residual in Each Barrier Subproblem')
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()
```
